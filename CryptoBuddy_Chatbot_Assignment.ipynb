Welcome to **CryptoBuddy**, your friendly chatbot that helps you explore cryptocurrency trends and sustainability.

```
# Step 1: Predefined Crypto Data
crypto_db = {
    "Bitcoin": {
        "price_trend": "rising",
        "market_cap": "high",
        "energy_use": "high",
        "sustainability_score": 3
    },
    "Ethereum": {
        "price_trend": "stable",
        "market_cap": "high",
        "energy_use": "medium",
        "sustainability_score": 6
    },
    "Cardano": {
        "price_trend": "rising",
        "market_cap": "medium",
        "energy_use": "low",
        "sustainability_score": 8
    }
}
```

```
# Step 2: Chatbot Logic and Functions

def greet_user():
    print("ðŸ‘‹ Hey there! Iâ€™m *CryptoBuddy*, your AI-powered crypto sidekick!")
    print("Ask me things like:")
    print("- Which crypto is trending up?")
    print("- Whatâ€™s the most sustainable coin?")
    print("- Tell me about Bitcoin\n")

def get_user_input():
    return input("ðŸ’¬ You: ").lower()

def process_input(user_input):
    if "trending" in user_input or "rising" in user_input:
        trending_cryptos = [name for name, data in crypto_db.items() if data["price_trend"] == "rising"]
        return f"ðŸ“ˆ These cryptos are currently trending up: {', '.join(trending_cryptos)}"

    elif "sustainable" in user_input or "most green" in user_input:
        most_sustainable = max(crypto_db.items(), key=lambda x: x[1]["sustainability_score"])
        return f"ðŸŒ± The most sustainable coin is *{most_sustainable[0]}* with a score of {most_sustainable[1]['sustainability_score']}/10."

    elif "bitcoin" in user_input:
        data = crypto_db["Bitcoin"]
        return format_crypto_data("Bitcoin", data)

    elif "ethereum" in user_input:
        data = crypto_db["Ethereum"]
        return format_crypto_data("Ethereum", data)

    elif "cardano" in user_input:
        data = crypto_db["Cardano"]
        return format_crypto_data("Cardano", data)

    elif "bye" in user_input or "exit" in user_input:
        return "ðŸ‘‹ Bye! Remember, always do your own research before investing. Stay crypto-curious! ðŸš€"

    else:
        return "ðŸ¤– Hmm... I didnâ€™t get that. Try asking about trends, sustainability, or a specific coin."

def format_crypto_data(name, data):
    return (f"ðŸ“Š Hereâ€™s the scoop on *{name}*:\n"
            f"- Price Trend: {data['price_trend'].capitalize()}\n"
            f"- Market Cap: {data['market_cap'].capitalize()}\n"
            f"- Energy Use: {data['energy_use'].capitalize()}\n"
            f"- Sustainability Score: {data['sustainability_score']}/10")

def run_chatbot():
    greet_user()
    while True:
        user_input = get_user_input()
        response = process_input(user_input)
        print(f"ðŸ¤– CryptoBuddy: {response}")
        if "bye" in user_input or "exit" in user_input:
            break
```

```
# Step 3: Run the Chatbot
run_chatbot()
```
